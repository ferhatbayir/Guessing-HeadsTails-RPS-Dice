Number guessing

```python
# We import the random module to generate random numbers
import random

# We set the range of possible numbers (1 to 100 in this case)
min_number = 1
max_number = 100

# We choose a random number within the range
secret_number = random.randint(min_number, max_number)

# We set the number of guesses allowed (3 in this case)
max_guesses = 5

# We print a welcome message and instructions
print("Welcome to the number guessing game!")
print(f"We are thinking of a number between {min_number} - {max_number}")
print(f"You have {max_guesses} guesses to find it.")

# We initialize a veriable to store the number of guesses made
guesses_made = 0

# We start a loop that repeats until the player runs out of guesses or the finds the secret number
while guesses_made < max_guesses:
    
    # Ask the player to enter a guess and convert it to the integer
    guess = int(input("Enter your guess: "))

    # Increment the number of guesses made by one
    guesses_made += 1

    # Chack if the guess is correct, too high or too low and print feedback
    if guess == secret_number:
        print("Congratulations! You got it! ")

    elif guess > secret_number:
        print("Your guess is too high.")

    else:
        print("Your guess is too low.")

    # We print how many guesses are left
    print(f"You have {max_guesses - guesses_made} guesses left.")

# If the loop ends without finding the secret number, print a losing message
else:
    print(f" Sorry, you run out of guesses. The secret number was {secret_number}")
```
