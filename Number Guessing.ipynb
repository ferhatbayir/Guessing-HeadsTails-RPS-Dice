Number guessing

```python
# We import the random module to generate random numbers
import random

# We set the range of possible numbers (1 to 100 in this case)
min_number = 1
max_number = 100

# We choose a random number within the range
secret_number = random.randint(min_number, max_number)

# We set the number of guesses allowed (3 in this case)
max_guesses = 5

# We print a welcome message and instructions
print("Welcome to the number guessing game!")
print(f"We are thinking of a number between {min_number} - {max_number}")
print(f"You have {max_guesses} guesses to find it.")

# We initialize a veriable to store the number of guesses made
guesses_made = 0

# We start a loop that repeats until the player runs out of guesses or the finds the secret number
while guesses_made < max_guesses:
    
    # Ask the player to enter a guess and convert it to the integer
    guess = int(input("Enter your guess: "))

    # Increment the number of guesses made by one
    guesses_made += 1

    # Chack if the guess is correct, too high or too low and print feedback
    if guess == secret_number:
        print("Congratulations! You got it! ")
        break

    elif guess > secret_number:
        print("Your guess is too high.")

    else:
        print("Your guess is too low.")

    # We print how many guesses are left
    print(f"You have {max_guesses - guesses_made} guesses left.")

# If the loop ends without finding the secret number, print a losing message
else:
    print(f" Sorry, you run out of guesses. The secret number was {secret_number}")
```

Heads-Tails Decider

```python
# We are import the random module the generate random numbers
import random

# We define a function named flip_coin
def flip_coin():
    # We choose one of the random values with the 'choice' function
    result = random.choice(["Heads", "Tails"])
    return result

# We create an inifite loop to print the values continiously on the screen
while True:
    value = flip_coin()
    print(value)

    # We add a command to stop the program at any time
    if input("Press 'q' to quit: ") == 'q':
        break

 
```

rock/paper/scissors Aplicaton

```python
import random

# We define a function named player_game
def play_game(player_choice):
    # We create a list and insert our required values
    choices = ["rock", "paper", "scissors"]
    # We make the computer choose randomly with the choice function 
    computer_choice = random.choice(choices)

    print("Player's choice:", player_choice)
    print("Computer's choice: ", computer_choice)

    # We create if blocks to determine the winner by comparing the results

    if player_choice == computer_choice:
        return "It's a tie!"
    
    elif (
        (computer_choice == "rock" and player_choice == "scissors") or
        (computer_choice == "paper" and player_choice == "rock") or
        (computer_choice == "scissors" and player_choice == "paper")
    ):
        return "Computer wins!"
    
    elif (
        (player_choice == "rock" and computer_choice == "scissors") or
        (player_choice == "paper" and computer_choice == "rock") or
        (player_choice == "scissors" and computer_choice == "paper")
    ):
        return "Player wins!"

    
    else:
        # We make a error shot
        raise ValueError("Ä°nvalid choice entered!")

try:
    # Player makes a selection and the input is converted to lowercase using the lower() method    
    player_choice = input("Enter your choice (rock/paper/scissors): ").lower()
    # play_game() function takes the player's choice as a parameter and assigns the result to the result variable
    result = play_game(player_choice)
    print(result)

except ValueError as e:
    print(e)
```
